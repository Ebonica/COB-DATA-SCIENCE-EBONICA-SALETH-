**CALLING PUBLIC API GIVEN AND CREATING A CSV DATASET USING PYTHON AND PANDAS API:**

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url = "https://data.binance.com/api/v3/ticker/24hr"

try:

    response = requests.get(url)

    if response.status_code == 200:

        data = response.json()
        df = pd.DataFrame(data)
        csv_filename = "binance_data.csv"
        df.to_csv(csv_filename, index=False)
        print(f"Data saved to {csv_filename}")
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

**PREPROCESSING THE DATASET**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
normalized_data=df.apply(LabelEncoder().fit_transform)
normalized_data.head()
```

```python
normalized_data.dtypes
```

**DATA VISUALIZATION**

```python
plt.figure(figsize=(10, 6))
plt.hist(normalized_data['lastPrice'], bins=50, color='skyblue')
plt.title('Distribution of Last Prices')
plt.xlabel('Last Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
N = 10
top_symbols = normalized_data.sort_values(by="volume", ascending=False).head(N)

plt.figure(figsize=(12, 6))
sns.barplot(x="volume", y="symbol", data=top_symbols, palette="viridis")
plt.xlabel("Volume")
plt.ylabel("Symbol")
plt.title(f"Top {N} Symbols by Trading Volume")
plt.tight_layout()
plt.show()
```

```python
symbol = normalized_data["symbol"]
last_price = normalized_data["lastPrice"]
volume = normalized_data["volume"]

plt.figure(figsize=(12, 6))
plt.bar(symbol, last_price)
plt.xlabel("Symbol")
plt.ylabel("Last Price")
plt.title("Last Prices for Binance Symbols")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(volume, last_price, alpha=0.5)
plt.xlabel("Volume")
plt.ylabel("Last Price")
plt.title("Volume vs. Last Price for Binance Symbols")
plt.grid(True)
plt.tight_layout()
plt.show()
```
