```python
nunique_counts = data.nunique()


for column in data.columns:
    unique_values = data[column].unique()
    value_counts = data[column].value_counts()

    print(f"Column: {column}")
    print(f"Number of unique values: {nunique_counts[column]}")
    print(f"Unique values: {unique_values}")
    print(f"Value counts:\n{value_counts}\n")
```

**DATA VISUALIZATION**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='type', data=data)
plt.xlabel('Type (Movie/TV Show)')
plt.ylabel('Count')
plt.title('Count of Movies vs. TV Shows')
plt.show()
```

```python
top_countries = data['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries with the Most Content')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='release_year', bins=30, kde=True)
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Distribution of Release Years')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='rating', data=data, order=data['rating'].value_counts().index)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Rating Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
pivot_table = data.pivot_table(index='rating', columns='type', values='show_id', aggfunc='count')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='g', cbar=True)
plt.xlabel('Type (Movie/TV Show)')
plt.ylabel('Rating')
plt.title('Count of Content by Rating and Type')
plt.show()
```

```python
top_directors = data['director'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.index, y=top_directors.values)
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top 10 Directors with the Most Content')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='release_year', hue='type')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Release Year vs. Count by Type')
plt.xticks(rotation=90)
plt.legend(title='Type', loc='upper right')
plt.show()
```

```python
categories = data['listed_in'].str.split(', ', expand=True).stack().value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=categories.values, y=categories.index, palette='viridis')
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('Count of Content in Different Categories')
plt.show()
```
